```python


resampled_week = pd.DataFrame(numeric.resample(rule='1W', on='날짜'))
target_columns = ['주식종가','변동금리','score1','score2']

week = []
for i in range(len(resampled_week)):
    temp = []
    for j in target_columns:
        temp.append(resampled_week[1][i].mean()[j])
    week.append(temp)
week_mean = pd.DataFrame(week)
```

```python
week_mean.rename({0:'주식종가',1:'변동금리',2:'score1',3:'score2'},axis=1,inplace=True)
```

```python
week_mean
```

```python
label = ['']
tp = list(week_mean['주식종가'])
for i in range(len(tp)-1):
    if tp[i] < tp[i+1]:
        label.append(1)
    else:
        label.append(0)
len(label)
```

```python
week_mean['label'] = label
```

```python
total_tbl_train = week_mean.loc[1:,['변동금리','score1','score2']]
total_tbl_label = week_mean.loc[1:,'label']
```

```python

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression # 앙상블 조합용
from sklearn.neighbors import KNeighborsClassifier # 앙상블 조합용
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
import xgboost as xgb

XGB = xgb.XGBClassifier()
gtb = GradientBoostingClassifier()
lr_clf = LogisticRegression()
knn_clf = KNeighborsClassifier(n_neighbors=4)
vo_clf = VotingClassifier(estimators=[('LR',lr_clf),('KN',knn_clf)],voting='hard')



X_train,X_test,y_train,y_test = train_test_split(
    total_tbl_train,total_tbl_label,test_size=0.2,random_state=10
)

```

```python
dt_clf = DecisionTreeClassifier()
rf_clf = RandomForestClassifier()
lr_clf = LogisticRegression()

estimator = [dt_clf,rf_clf,lr_clf,knn_clf,vo_clf,gtb,XGB]

# df_clf.fit(x_train,y_train)
# pred = df_clf.predict(x_test)
# accuracy_score(y_test,pred)
def bundle(data,label,rate,estimator):
    X_train,X_test,y_train,y_test = train_test_split(total_tbl_train,total_tbl_label,test_size=rate,random_state=10)
    
    cv_acc = []
    for clf in estimator:
        clf.fit(X_train, y_train)  # classifier
        pred = clf.predict(X_test)

        acc = np.round(accuracy_score(y_test, pred), 4)
        cv_acc.append(acc)

    return cv_acc
```

```python
from sklearn.model_selection import GridSearchCV
# parms = {'n_estimators':[],'max_depth':[],'min_samples_split':[],'min_samples_leaf':[]}
parms = {'n_estimators':list(range(30,80,5)), # num of iter : using classifier
         'max_depth':list(range(6,15,3)),
         'min_samples_split':list(range(1,15,2)),
         'min_samples_leaf':list(range(2,8,2))}
rf_clf = RandomForestClassifier() # n_jobs ?
grid_cv = GridSearchCV(rf_clf,param_grid=parms,cv=5,n_jobs=-1,)
grid_cv.fit(X_train,y_train)
print(grid_cv.best_params_)
print(round(grid_cv.best_score_,4))

grid_cv.best_estimator_
```
