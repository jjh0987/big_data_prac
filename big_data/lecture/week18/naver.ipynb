```python
import pandas as pd
import numpy as np
import requests
import re
from bs4 import BeautifulSoup

import warnings
warnings.filterwarnings('ignore')
from transformers import pipeline
qa = pipeline('question-answering')
zs = pipeline('zero-shot-classification')
```

```python
df = pd.read_csv('/Users/junho/Downloads/naver기사_전처리_번역_score1&2.csv').drop('Unnamed: 0',axis=1)
```

```python
translated = df['번역']
```

```python
score2 = []
question2 = f'Can NAVER Grow in the Future?'
label2 = ['grow POSITIVE', 'grow NEGATIVE']
cnt = 0
fail = 0
marking = []
length = len(translated)
```

```python
for sequence in list(translated):
    try:
        ans = qa(question = question2, context = sequence)
        temp = zs(ans.get('answer'),label2)
        temp_score = abs(temp['scores'][0]-temp['scores'][1])
        
        cnt += 1

        if temp['labels'][0] == 'grow NEGATIVE':
            score2.append(-temp_score)
            print(f'{cnt}/{length}' + f': append -{temp_score}')
        else:
            score2.append(temp_score)
            print(f'{cnt}/{length}' + f': append {temp_score}')
    except:
        score2.append('')
        cnt += 1
        print('except')
```
