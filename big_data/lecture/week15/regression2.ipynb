```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./sample1.csv')
df
```

```python
def averaging(column):
    section1 = []
    section2 = []
    section3 = []
    for i in range(len(df)):
        if i <= 24:
            section1.append(df[column][i])
        elif i <= 38:
            section2.append(df[column][i])
        else:
            section3.append(df[column][i])
    df[column][:25] = np.mean(section1)
    df[column][25:39] = np.mean(section2)
    df[column][39:] = np.mean(section3)
    return df

averaging('testimony')
averaging('minutes')
averaging('beige')
averaging('speech1')
averaging('speech2')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

```python
input_idx=45

lr = LinearRegression()
poly = PolynomialFeatures(degree=2, include_bias=False)
x = df.loc[:,['testimony', 'minutes', 'beige', 'speech1', 'speech2']]
y = df.loc[:, 'FEDfunds']
x_poly = poly.fit_transform(x)
lr.fit(x_poly, y)
y_pred = lr.predict(poly.fit_transform(x))

pred = []
for i in y_pred:
    if i < 0:
        pred.append(0)
    else:
        pred.append(i)

len(pred)

# columns : 'testimony', 'minutes', 'beige', 'speech1', 'speech2', 'btc', 'inf', 'uem', 'label'
```

```python
rmse = mean_squared_error(y, y_pred, squared=False)
print(rmse)
```
