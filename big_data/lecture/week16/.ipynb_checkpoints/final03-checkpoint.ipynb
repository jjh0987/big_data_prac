```python
import pandas as pd
stock_datgle = pd.read_csv('/Users/junho/Downloads/네이버종토방댓글.csv')
```

```python
stock_datgle['날짜'] = pd.to_datetime(stock_datgle['날짜'],format='%Y-%m-%d')
stock_datgle
```

```python
temp = pd.DataFrame(stock_datgle.groupby('날짜'))
stock_datgle = pd.DataFrame(temp.loc[len(temp)-365:len(temp)])
```

```python
from googletrans import Translator
translator = Translator()
ex = translator.translate('오늘은 날씨가 좋습니다. 내일도 날씨가 좋을까요?', src = 'ko', dest = 'en')
```

```python
ex.text
```

```python
import time
translated_datgle = []
for i in list(stock_datgle.iloc[0,1]['댓글']):
    temp = translator.translate(i).text
    time.sleep(1)
    print(temp)
    translated_datgle.append(temp)
```

```python
pd.DataFrame(translated_datgle).to_csv('/Users/junho/Desktop/data/final_csv/sample_translated_datgle.csv')
```
