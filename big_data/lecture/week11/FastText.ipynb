## 1.2 preprocessing

```python
import re

def find_hangul(text):
    return re.findall(f'[ㄱ-ㅎ가-힣]+',text)
```

```python
sub_df = nsmc['document']
sub_df
```

```python
nsmc.iloc[0,1]
```

```python
find_hangul(nsmc.iloc[0,1])
```

```python
nsmc[nsmc['document'].isnull()]
```

```python
nsmc[nsmc['document'].notnull()]
```

```python
data = nsmc[nsmc['document'].notnull()]['document'].map(find_hangul) # map : 함수 바로적용
data
```

```python
def only_hangul(text):
    return ' '.join(find_hangul(text))

only_hangul(nsmc.loc[0,'document']) # find_hangul -> join 이므로 nsmc 가져와야함.
```

```python
data2 = pd.DataFrame(nsmc[nsmc['document'].notnull()]['document'].map(only_hangul)) # map : 함수 바로적용
data2
```

```python
data2.to_csv('nsmc.txt',sep='\t',encoding='utf8')
```

```python
with open('nsmc.txt','w',encoding='utf8') as f:
    f.write('\n'.join(data2))
```

## 1.3 fast text : n-gram tokenization

```python
from gensim.models.word2vec import Word2Vec
from gensim.models.fasttext import FastText
```

```python
model = FastText(vector_size=16) 
# imbeddig size #option, sg=0 : cbow, sg=1 : skip-gram
# alpha = 0.025 : default learning rate
# min_alpha : 학습률을 점점 낮추는 정도
# window = 5 : 좌우 학습단어 갯수 총 2*window 개
# min_count = 5 : 최소 min_count 만큼 있어야 학습
# vector_size = 100
```

```python
model.build_vocab(corpus_iterable=data)
```

```python
data
```

```python
model.build_vocab(corpus_file='nsmc.txt')
```

```python
model.train(
    corpus_iterable=data,
    epochs=5,
    total_examples=model.corpus_count,
    total_words=model.corpus_total_words
)
```
