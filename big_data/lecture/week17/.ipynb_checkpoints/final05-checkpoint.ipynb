```python
numeric = pd.read_csv('/Users/junho/Downloads/VIX-나스닥-주가-금리.csv')[76:].drop('Unnamed: 0',axis=1)
#numeric.drop('날짜',axis=1,inplace=True)
#numeric.reset_index().drop('index',axis=1) # 21/4/28 - 22/4/28
numeric['날짜'] = [i.replace('년 ','-').replace('월 ','-').replace('일','') for i in list(numeric['날짜'])]
numeric['날짜'] = pd.to_datetime(numeric['날짜'],format='%Y-%m-%d')
numeric = numeric.reset_index().drop('index',axis=1)
numeric
```

```python
numeric = pd.merge(numeric,pykrx.loc[:,['날짜','label']],on='날짜').drop('주식종가',axis=1)
```

```python
numeric.info()
```

```python
drop_target = ['Unnamed: 0','Unnamed: 0.1','제목','영문번역']
score_tbl = pd.read_csv('/Users/junho/Downloads/카카오뉴스기사_최종.csv').drop(drop_target,axis=1)
score_tbl['날짜'] = pd.to_datetime(score_tbl['날짜'],format='%Y-%m-%d')
score_tbl = pd.DataFrame(score_tbl.groupby('날짜').mean())
score_tbl.reset_index(inplace=True)
score_tbl
```

```python
kospi = pd.read_csv('/Users/junho/Downloads/코스피+코스닥지수 내역.csv',encoding='cp949')
kospi['날짜'] = [i.replace('년 ','-').replace('월 ','-').replace('일','') for i in list(kospi['날짜'])]
kospi['날짜'] = pd.to_datetime(kospi['날짜'],format='%Y-%m-%d')
kospi
```

```python
total_tbl = pd.merge(numeric,score_tbl,on='날짜')
total_tbl = pd.merge(kospi.loc[:,['날짜','종가','종가.1']],total_tbl,on='날짜')
total_tbl_train = total_tbl.loc[:,['vix종가','나스닥종가','변동금리','fear/greed','price_rise/price_drop']]
total_tbl_label = total_tbl.label
total_tbl # 종가 : 코스피 , 종가.1 : 코스닥
```

```python
total_tbl.info()
```

```python
#total_tbl.drop('날짜',axis=1,inplace=True)
#total_tbl['나스닥종가'] = [i.replace(',','') for i in total_tbl['나스닥종가']]
#total_tbl = total_tbl.astype('float') 
```

```python
total_tbl_train = total_tbl.loc[:,['vix종가','변동금리','price_rise/price_drop']]
#total_tbl_train['나스닥종가'] = [i.replace(',','') for i in total_tbl_train['나스닥종가']]
total_tbl_train = total_tbl_train.astype('float') 
total_tbl_train.info()
```

```python
import pandas as pd
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression # 앙상블 조합용
from sklearn.neighbors import KNeighborsClassifier # 앙상블 조합용
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

lr_clf = LogisticRegression()
knn_clf = KNeighborsClassifier(n_neighbors=8)
vo_clf = VotingClassifier(estimators=[('LR',lr_clf),('KN',knn_clf)],voting='soft')



X_train,X_test,y_train,y_test = train_test_split(
    total_tbl_train,total_tbl_label,test_size=0.2,random_state=10
)
```

```python
X_train.info()
```

# voting

```python
total_tbl_train
```

```python
vo_clf.fit(X_train,y_train)
```

```python
accuracy_score(vo_clf.predict(X_test),y_test)
```

# random forest

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier(random_state=0)
rf_clf.fit(X_train,y_train)
pred = rf_clf.predict(X_test)
acc = accuracy_score(pred,y_test)
acc
```

```python
from sklearn.model_selection import GridSearchCV
# parms = {'n_estimators':[],'max_depth':[],'min_samples_split':[],'min_samples_leaf':[]}
parms = {'n_estimators':list(range(30,80,5)), # num of iter : using classifier
         'max_depth':list(range(9,15,3)),
         'min_samples_split':list(range(1,10,2)),
         'min_samples_leaf':list(range(2,8,2))}
rf_clf = RandomForestClassifier(random_state=0,n_jobs=-1) # n_jobs ?
grid_cv = GridSearchCV(rf_clf,param_grid=parms,cv=5,n_jobs=-1,)
grid_cv.fit(X_train,y_train)
print(grid_cv.best_params_)
print(round(grid_cv.best_score_,4))
```

# Gradient boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
gtb = GradientBoostingClassifier(max_depth=5)
gtb.fit(X_train,y_train)
acc = accuracy_score(gtb.predict(X_test),y_test)
acc
```
