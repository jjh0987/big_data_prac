```python
import pandas as pd
finance_art1 = pd.read_csv('/Users/junho/Desktop/data/final_csv/파이낸셜뉴스_translated_article.csv')
```

```python
finance_art1 = finance_art1.drop('Unnamed: 0',axis=1)
```

```python
finance_art1
```

```python
from transformers import pipeline
zs = pipeline('zero-shot-classification')
```

```python
score = []
label = ['fear','greed']
cnt = 0
for sequence in list(finance_art1['0']):
    temp = zs(sequence,label)
    temp_score = abs(temp['scores'][0]-temp['scores'][1])
    cnt += 1

    if temp['labels'][0] == 'fear':
        score.append(-temp_score)
        print(f'{cnt}/{len(finance_art1)}' + f': append -{temp_score}')
    else:
        score.append(temp_score)
        print(f'{cnt}/{len(finance_art1)}' + f': append {temp_score}')
```

```python
score_df = pd.DataFrame(data=score,columns=['fear/greed idx'])
score_df.plot()
```
