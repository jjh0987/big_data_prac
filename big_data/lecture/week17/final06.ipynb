# 정형 데이터

```python
numeric = pd.read_csv('/Users/junho/Desktop/data/final_csv/VIX-나스닥-주가-금리.csv')[77:].drop('Unnamed: 0',axis=1)
numeric
```

```python
temp = pd.read_csv('/Users/junho/Desktop/data/final_csv/코스피+코스닥지수 내역.csv',encoding='cp949')
temp.loc[:,['날짜','종가','종가.1']]
```

```python
total = pd.merge(numeric,temp.loc[:,['날짜','종가','종가.1']],on='날짜')
total['나스닥종가'] = [i.replace(',','') for i in list(total['나스닥종가'])]
total['코스피'] = [i.replace(',','') for i in list(total['종가'])]
total['코스닥'] = [i.replace(',','') for i in list(total['종가.1'])]

total.drop(['종가','종가.1'],axis=1,inplace=True)
total['날짜'] = [i.replace('년 ','-').replace('월 ','-').replace('일','') for i in list(total['날짜'])]
total
```

```python
total['나스닥종가'] = total['나스닥종가'].astype('float')
total['코스닥'] = total['코스닥'].astype('float')
total['코스피'] = total['코스피'].astype('float')
total['날짜'] = pd.to_datetime(total['날짜'],format='%Y-%m-%d')
```

```python
total.to_csv('/Users/junho/Desktop/data/final_csv/종합지표.csv')
total.info()
```
