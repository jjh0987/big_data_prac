```python

resampled_week = pd.DataFrame(numeric.resample(rule='1D', on='날짜'))
target_columns = ['주식종가','변동금리','score1','score2','score3','개인','기관 합계','외국인 합계','vix종가']

week = []
for i in range(len(resampled_week)):
    temp = []
    for j in target_columns:
        temp.append(resampled_week[1][i].mean()[j])
    week.append(temp)
week_mean = pd.DataFrame(week)
```

```python
week_mean.rename({0:'주식종가',1:'변동금리',2:'score1',3:'score2',4:'score3',5:'개인',6:'기관 합계',7:'외국인 합계',8:'vix종가'},axis=1,inplace=True)
```

```python
week_mean.dropna(inplace=True)
```

```python
label = ['']
tp = list(week_mean['주식종가'])
for i in range(len(tp)-1):
    if tp[i] < tp[i+1]:
        label.append(1)
    else:
        label.append(0)
len(label)
```

```python
week_mean['label'] = label
```

```python
week_mean = week_mean.loc[1:,:]
```

```python
week_mean['label'] = week_mean['label'].astype('float64')
```

```python
week_mean
```

```python
total_tbl_train = week_mean.loc[1:,['score1','score2','score3']]
total_tbl_label = week_mean.loc[1:,'label']
```

```python
from sklearn.model_selection import GridSearchCV

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression # 앙상블 조합용
from sklearn.neighbors import KNeighborsClassifier # 앙상블 조합용
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
import xgboost as xgb

lr_clf = LogisticRegression(random_state=10)
XGB = xgb.XGBClassifier(random_state=10)
gtb = GradientBoostingClassifier(random_state=10)
knn_clf = KNeighborsClassifier(n_neighbors=4)
vo_clf = VotingClassifier(estimators=[('LR',lr_clf),('KN',knn_clf)],voting='hard')
dt_clf = DecisionTreeClassifier(random_state=10)
rf_clf = RandomForestClassifier(random_state=10)

import warnings
warnings.filterwarnings('ignore')

X_train,X_test,y_train,y_test = train_test_split(
    total_tbl_train,total_tbl_label,test_size=0.25,random_state=10
)


estimator = [dt_clf,rf_clf,lr_clf,knn_clf,vo_clf,gtb,XGB]


def bundle(data,label,rate,estimator,random):
    X_train,X_test,y_train,y_test = train_test_split(total_tbl_train,total_tbl_label,test_size=rate,random_state=random)
    
    cv_acc = []
    
    for clf in estimator:
        '''
        temp = []
        if str(clf) == 'RandomForestClassifier(random_state=10)':
            parms = {'n_estimators':list(range(30,80,5)), # num of iter : using classifier
                     'max_depth':list(range(6,15,3)),
                     'min_samples_split':list(range(1,15,2)),
                     'min_samples_leaf':list(range(2,8,2))}
            grid_cv = GridSearchCV(clf,param_grid=parms,cv=5,n_jobs=-1,)
            grid_cv.fit(X_train,y_train)

            md = grid_cv.best_estimator_
            md.fit(X_train,y_train)

            pred = grid_cv.best_estimator_.predict(X_test)

            acc = np.round(accuracy_score(y_test, pred), 4)
            cv_acc.append(acc)
            
        else:
        '''
        clf.fit(X_train, y_train)  # classifier
        pred = clf.predict(X_test)

        acc = np.round(accuracy_score(y_test, pred), 4)
        cv_acc.append(acc)
        
        
    return cv_acc
```

```python
import numpy as np
print(estimator)
print('-'*50)

for i in range(6):
    acc = bundle(total_tbl_train,total_tbl_label,0.2+0.02*i,estimator,36)
    print(acc)

    vals = []
    for j in estimator:
        val = j.predict(np.array(X_test.iloc[-1,:]).reshape(1, -1))
        vals.append(val)

    check = vals[np.argmax(acc)]

    print([int(i) for i in vals])
    if int(check) == 1:
        print(f'다음날 주가는 상승할것으로 예측됩니다. 가장 높은 예측 정확도는 {round(max(acc)*100,2)}% 입니다.')
    elif int(check) == 0:
        print(f'다음날 주가는 하락할것으로 예측됩니다. 가장 높은 예측 정확도는 {round(max(acc)*100,2)}% 입니다.')

    print()
```
